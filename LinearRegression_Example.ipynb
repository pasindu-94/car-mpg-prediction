#Model Training and Development

Model Function : y=f(X)


X = [x1,x2,x3...]

Target or Label : Y

Model Function : y = f([x1, x2, x3..])

##Model Development done using Scikit Learn

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
```

```python
print(model)
```

```python
X = ['weight']
y_actual = 'mpg'
y_predict = 'predicted_mpg'
```

#Train the Model and Predict

```python
model.fit(data[X], data[y_actual])
```

```python
coefficeients = model.coef_
intercept = model.intercept_

print("Coefficeient =" + str(coefficeients))
print("Intercept =" + str(intercept))
```

##Prediction

```python
test_weight = 2500.00
```

```python
model.predict([[test_weight]])
```

```python
data[y_predict] = model.predict(data[X])
```

```python
data.head()
```

```python
data['difference'] = data[y_actual] - data[y_predict]
```

```python
data.head()
```

```python
data[[y_actual, y_predict, 'difference']]
```

#Evaluat the Model

MSE : Mean Square Error
RMSE : Root Mean Square Error
R2 : Coefficient of Determination

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(data[y_actual], data[y_predict])
print("MSE :" ,mse)

rmse = np.sqrt(mse)
print("RMSE :" ,rmse)

r2 = r2_score(data[y_actual], data[y_predict])
print("R2 :" , r2)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(data[y_actual], data[y_predict])
plt.xlabel('mpg actual')
plt.ylabel('mpg predicted')
plt.title('Car MPG Prediction')
```
